```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--disable-popup-blocking')
chrome_options.add_argument('--profile-directory=Default')
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--disable-plugins-discovery')
chrome_options.add_argument('--no-first-run')
chrome_options.add_argument('--no-service-autorun')
chrome_options.add_argument('--no-default-browser-check')
chrome_options.add_argument('--password-store=basic')
chrome_options.add_argument('--no-sandbox')


driver = webdriver.Chrome(options=chrome_options)
```

```python
url = "https://weibo.com/5709745316?refer_flag=1001030103_"
driver.get(url)
```

```python
fens_xp = "//div//span[@class,'粉丝')]"
guanz_xp = "//div/span[@class,'关注')]"
gender_xp = "//div[@class,'ProfileHeader_h3_2nhjc']//span"
ip_adress_xp = "//div[@class,'IP属地']"
```

```python
df_weibo_user_info = pd.read_csv(r"核污水微博贴文原始数据-数据清理.csv")
df_weibo_user_info
```

```python
df_all = pd.DataFrame()
for index_,row in df_weibo_user_info.iterrows():
    user_name = row['user_name']
    user_name_href = row['user_name_href']
    print(index_,user_name_href)
    try:
        driver.get(user_name_href)
    except:
        try:
            driver.get(user_name_href)
        except:
            print("用户主页无法访问")
            break

    try:
        WebDriverWait(driver,15).until(EC.visibility_of_element_located((By.XPATH,"//div[contains(@class,'ProfileHeader_name')]")))
    except:
        print('用户信息显示错误')
        break

    time.sleep(1)

    try:
        fens = driver.find_element(By.XPATH,fens_xp).get_attribute('textContent').replace('粉丝','').strip()
    except:
        fens = ''

    try:
        guanz = driver.find_element(By.XPATH,guanz_xp).get_attribute('textContent').replace('关注','').strip()
    except:
        guanz = ''

    try:
        gender = driver.find_element(By.XPATH,gender_xp).get_attribute('title').strip()
    except:
        gender = ''

    try:
        ip_adress = driver.find_element(By.XPATH,ip_adress_xp).get_attribute('textContent').replace('IP属地：','').strip()
    except:
        ip_adress = ''

    df_weibo_user = pd.DataFrame({"user_name":[user_name],"user_name_href":[user_name_href],"fens":[fens],"guanz":[guanz],"gender":[gender],"ip_adress":[ip_adress]})

    df_all = pd.append(df_weibo_user)



```

```python
df_all.to_csv("微博用户原始信息.csv", index=False, encoding='utf-8-sig')
```
