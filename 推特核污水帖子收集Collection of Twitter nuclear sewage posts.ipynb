```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--disable-popup-blocking')
chrome_options.add_argument('--profile-directory=Default')
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--disable-plugins-discovery')
chrome_options.add_argument('--no-first-run')
chrome_options.add_argument('--no-service-autorun')
chrome_options.add_argument('--no-default-browser-check')
chrome_options.add_argument('--password-store=basic')
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get("https://twitter.com/search?q=nuclear%20contaminated%20water%20lang%3Aen%20until%3A2023-09-21%20since%3A2023-07-21&src=typed_query&f=live")
```

```python
blocks_xpath = "//div[@data-testid='tweet']"
user_name_xpath = "//span[text()='@']"
concent_xpath = "/div[@data-testid='tweetText']"
date_xpath = "//time"
post_href_xpath = "//time/a"
reply_xp = "//div[@data-testid='reply']"
retweet_xp = "//div[@data-testid='retweet']"
like_xpath = "//div[@data-testid='like']"
fenxi_xp = "//a[@aria-label='分析']"
```

```python
df_all = pd.DataFrame()
for block in driver.find_elements(By.XPATH,blocks_xpath):
    try:
        date_ = block.find_element(By.XPATH,date_xpath).get_attribute('datetime')
    except:
        date_ = ''
    try:
        name_ = block.find_element(By.XPATH,user_name_xpath).get_attribute('textContent')
    except:
        name_ = ''
    try:
        concent_ = block.find_element(By.XPATH,concent_xpath).get_attribute('textContent')
    except:
        concent_ = ''

    try:
        post_href_ = block.find_element(By.XPATH,post_href_xpath).get_attribute('href')
    except:
        try:
            post_href_ = block.find_element(By.XPATH,fenxi_xp).get_attribute('href').replace('/analytics','').strip()
        except:
            post_href_ = ''

    try:
        reply_ = block.find_element(By.XPATH,reply_xp).get_attribute('textContent')
    except:
        reply_ = ''

    try:
        retweet_ = block.find_element(By.XPATH,retweet_xp).get_attribute('textContent')
    except:
        retweet_ = ''

    try:
        like_ = block.find_element(By.XPATH,like_xpath).get_attribute('textContent')
    except:
        like_ = ''

    try:
        fenxi_ = block.find_element(By.XPATH,fenxi_xp).get_attribute('textContent')
    except:
        fenxi_ = ''
    

    data_dic = [{"date_": date_,"name_":name_,"concent_":concent_,"post_href_":post_href_,"reply_":reply_,"retweet_":retweet_,"like_":like_,"fenxi_":fenxi_}]
    df_datas = pd.DataFrame(data_dic)
    df_all = pd.concat([df_all,df_datas],ignore_index=True)
```

```python
df_all.to_csv(r"NuclearContaminatedWaterFromTwitter.csv",index=False,encoding='utf-8_sig')
```

```python
df_all_user = pd.DataFrame()
for index_,row in df_all_weibo_twutter_user_copy.iterrows():
    print(index_)
    name_ = row['name_']
    user_url = "https://twitter.com/{}".format(name_)
    print(user_url)
    try:
        driver.get(user_url)
    except:
        try:
            driver.get(user_url)
        except:
            print('访问用户主页失败')
            break
    
    try:
        WebDriverWait(driver,10).until(EC.visibility_of_element_located((By.XPATH,user_name_xp)))
    except:
        print("页面显示错误")
        break
    
    try:
        user_desc = driver.find_element(By.XPATH,user_desc_xp).get_attribute('textContent').strip()
    except:
        user_desc = ''
        
    try:
        user_join_date = driver.find_element(By.XPATH,user_join_date_xp).get_attribute('textContent').strip()
    except:
        user_join_date = ''
        
    try:
        user_followers_num = driver.find_element(By.XPATH,user_followers_num_xp).get_attribute('textContent').strip()
    except:
        user_followers_num = ''
        
    try:
        user_location = driver.find_element(By.XPATH,user_location_xp).get_attribute('textContent').strip()
    except:
        user_location = ''

    
    
    data_dic = [{"name_": name_,"user_url":user_url,"user_desc":user_desc,"user_join_date":user_join_date,"user_followers_num":user_followers_num,"user_location":user_location}]
    df_datas = pd.DataFrame(data_dic)
    df_all_user = pd.concat([df_all_user,df_datas])
    
```
