```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument('--disable-popup-blocking')
chrome_options.add_argument('--profile-directory=Default')
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--disable-plugins-discovery')
chrome_options.add_argument('--no-first-run')
chrome_options.add_argument('--no-service-autorun')
chrome_options.add_argument('--no-default-browser-check')
chrome_options.add_argument('--password-store=basic')
chrome_options.add_argument('--no-sandbox')


driver = webdriver.Chrome(options=chrome_options)
```

```python
block_eles_xp = "//div[@id='pl_feedlist_index']"
user_name_xp = ".//a[@class='name']"
zhankai_xp = ".//div[@class='content']"
user_self_content_xp = ".//div[@class='content']"
user_self_zhuanfa_num_xp = ".//div[@class='card-act']"
user_self_pl_num_xp = ".//div[@class='card-act']"
user_self_like_num_xp = ".//div[@class='card-act']"
user_post_datetime_xp = ".//div[@class='from']"
next_page_xp = "//div[@class='m-page']"
```

```python
import datetime

# 
start_date = datetime.date(2023, 7, 21)
end_date = datetime.date(2023, 9, 21)

# URL
urls = []

# 
current_date = start_date
while current_date <= end_date:
    # 151
    for page in range(1, 52):
        url = f"https://s.weibo.com/weibo?q=%E6%A0%B8%E6%B1%A1%E6%B0%B4&typeall=1&suball=1&timescope=custom:{current_date}:{current_date}&Refer=g&page={page}"
        urls.append(url)
    current_date += datetime.timedelta(days=1)

# URL
for url in urls:
    print(url)
```

```python
df_all = pd.DataFrame()
for url_li in url:
    try:
        driver.get(url_li)
    except:
        try:
            driver.get(url_li)
        except:
            print("")
            break
    while True:

        try:
            WebDriverWait(driver,5).until(EC.visibility_of_element_located((By.XPATH,block_eles_xp)))
        except:
            print("")
            break
        if driver.current_url.split('page=')[-1] == '51':break
        
        block_eles = driver.find_elements(By.XPATH,block_eles_xp)
        for block_ele in block_eles:
        
            try:
                block_ele.find_element(By.XPATH,zhankai_xp).click()
                time.sleep(0.3)
            except:
                pass
            
            
            try:
                user_name = block_ele.find_element(By.XPATH,user_name_xp).get_attribute('textContent').strip()
            except:
                user_name = ''
            
            try:
                user_name_href = block_ele.find_element(By.XPATH,user_name_xp).get_attribute('href').strip()
            except:
                user_name_href = ''
                
            try:
                user_self_content = block_ele.find_element(By.XPATH,user_self_content_xp).get_attribute('textContent').strip()
            except:
                user_self_content = ''
                
            try:
                user_self_zhuanfa_num = block_ele.find_element(By.XPATH,user_self_zhuanfa_num_xp).get_attribute('textContent').replace('','').strip()
            except:
                user_self_zhuanfa_num = ''
                
            try:
                user_self_pl_num = block_ele.find_element(By.XPATH,user_self_pl_num_xp).get_attribute('textContent').replace('','').strip()
            except:
                user_self_pl_num = ''
                
            try:
                user_self_like_num = block_ele.find_element(By.XPATH,user_self_like_num_xp).get_attribute('textContent').replace('','').strip()
            except:
                user_self_like_num = ''
                
            try:
                user_post_datetime = block_ele.find_element(By.XPATH,user_post_datetime_xp).get_attribute('textContent').strip()
            except:
                user_post_datetime = ''
            
            try:
                user_post_href = block_ele.find_element(By.XPATH,user_post_datetime_xp).get_attribute('href').strip()
            except:
                user_post_href = ''
            
            df = pd.DataFrame({"user_name":[user_name],"user_name_href":[user_name_href],"user_self_content":[user_self_content],"user_self_zhuanfa_num":[user_self_zhuanfa_num],"user_self_pl_num":[user_self_pl_num],"user_self_like_num":[user_self_like_num],"user_post_datetime":[user_post_datetime],"user_post_href":[user_post_href]})
            
            
            df_all.append(df)
        
        try:
            driver.find_element(By.XPATH,next_page_xp).click()
        except:
            print("")
            break
```

```python
df_all.to_csv("")
```
