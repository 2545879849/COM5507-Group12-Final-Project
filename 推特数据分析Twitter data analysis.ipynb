```python
from pyecharts import options as opts
from pyecharts.charts import WordCloud as EchartsWordCloud
from collections import Counter
import nltk
```

```python
# 读取停用词表
with open(r'stopwords-en.txt', 'r', encoding='utf-8') as f:
    stop_words = set(f.read().splitlines())
```

```python
extra_stop_words = [
    'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
    'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z',
    ',', '.', '!', '?', ';', ':', '-', '_', '+', '=', '@', '#', '$',
    '%', '^', '&', '*', '(', ')', '[', ']', '{', '}', '<', '>', '/',
    '\\', '|', '`', '~', "'", '"','shoes','shoe','ve','re','lo','pe','id','na','yr','UR','I0','|0','wo','|0'
]

stop_words.update(extra_stop_words)

# 定义分词函数
def tokenize(text):
    words = nltk.word_tokenize(text)
    return [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
```

```python
df_twitter_post['tokenized_content'] = df_twitter_post['concent_'].apply(tokenize)
df_twitter_post
```

```python
all_text = ' '.join(df_twitter_post['tokenized_content'].sum())
```

```python
wordcloud = EchartsWordCloud()
wordcloud.add("", [(word, all_text.count(word)) for word in set(all_text.split())], word_size_range=[20, 100])
wordcloud.set_global_opts(title_opts=opts.TitleOpts(title='Twitter核污水词云图'))
remove_words = {'a', 'b', 'c','t','b','k','h','|','/','n','→','↑','【','】','~','(',')','!','！','t','s','←','=','#','...','.','·','N','Japan','contaminated','日本','nuclear','water','Japan','contaminated','nuclear','water','nuclear contaminated water'
,'Nuclear wastewater'}
stop_words.update(remove_words)
wordcloud.render("Twitter_wordcloud.html")
```

```python
from snownlp import SnowNLP
import pandas as pd

# 假设 df_all_weibo_content 是你的数据帧，'微博内容' 是微博内容列
df_twitter_post['score'] = df_twitter_post['concent_'].apply(lambda x: SnowNLP(x).sentiments)


# 定义判断情感标签的函数
def get_sentiment_label(score, threshold_pos=0.6, threshold_neg=0.4):
    if score > threshold_pos:
        return 'positive'
    elif score < threshold_neg:
        return 'negative'
    else:
        return 'neutral'
df_twitter_post['label'] = df_twitter_post['score'].apply(get_sentiment_label)
df_twitter_post
```

```python
df_twitter_post
df_twitter_post.to_csv("df_twitter_post.csv", index=False, encoding='utf-8-sig')
```

```python
from pyecharts.charts import Bar, Pie
from pyecharts import options as opts
import pandas as pd
```

```python
# 统计各标签的行数
label_counts = df_twitter_post['label'].value_counts()

# 绘制柱状图
bar_chart = (
    Bar()
    .add_xaxis(label_counts.index.tolist())
    .add_yaxis("情感标签", label_counts.values.tolist())
    .set_global_opts(title_opts=opts.TitleOpts(title="Twitter推文情感标签统计"))
)

# 绘制扇形图
pie_chart = (
    Pie()
    .add("", [list(z) for z in zip(label_counts.index.tolist(), label_counts.values.tolist())])
    .set_global_opts(title_opts=opts.TitleOpts(title="Twitter推文情感标签比例"))
    .set_series_opts(
        radius=["30%", "75%"],
        center=["50%", "50%"],
        rose_type="radius",
        label_opts=opts.LabelOpts(is_show=True),
    )
)

# 生成 HTML 文件
bar_chart.render("emotion_label_bar_chart.html")
pie_chart.render("emotion_label_pie_chart.html")
```

```python
import pandas as pd
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
df_twitter_post['date_'] = pd.to_datetime(df_twitter_post['date_'])

# 统计每天发帖数量
daily_post_count = df_twitter_post.groupby(df_twitter_post['date_'].dt.date)['date_'].count()
```

```python
# 绘制折线图
line_chart = (
    Line()
    .add_xaxis(daily_post_count.index.tolist())
    .add_yaxis("每天发帖量", daily_post_count.values.tolist(), is_smooth=True, markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Twitter发帖量统计"),
        xaxis_opts=opts.AxisOpts(name="日期"),
        yaxis_opts=opts.AxisOpts(name="发帖量"),
        legend_opts=opts.LegendOpts(pos_right="10%"),
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),  # 隐藏标签
        line_style_opts=opts.LineStyleOpts(width=5),  # 折线粗细
    )
)

line_chart.set_series_opts(
    label_opts=opts.LabelOpts(is_show=False),
    linestyle_opts=opts.LineStyleOpts(width=2),  # 折线粗细
)

# 调整图形大小
line_chart.width = "80%"
line_chart.height = "500px"

# 生成 HTML 文件
line_chart.render("daily_post_count_line_chart.html")
```

```python
# 统计每天发帖数量和情感分数平均值

daily_average_score = df_twitter_post.groupby(df_twitter_post['date_'].dt.date)['score'].mean()
# 绘制折线图
line_chart = (
    Line()
    .add_xaxis(daily_average_score.index.tolist())
    .add_yaxis("均值", daily_average_score.values.tolist(), is_smooth=True, markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Twitter推文情感趋势图"),
        xaxis_opts=opts.AxisOpts(name="日期"),
        yaxis_opts=opts.AxisOpts(name="情绪分值"),
        legend_opts=opts.LegendOpts(pos_right="10%"),
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),  # 隐藏标签
        line_style_opts=opts.LineStyleOpts(width=5),  # 折线粗细
    )
)

line_chart.set_series_opts(
    label_opts=opts.LabelOpts(is_show=False),
    linestyle_opts=opts.LineStyleOpts(width=2),  # 折线粗细
)

# 调整图形大小
line_chart.width = "80%"
line_chart.height = "500px"

# 生成 HTML 文件
line_chart.render("sentiment_trend_line_chart.html")

```
