```python
import re
def convert_to_number(value):
    if pd.isna(value) or value == '':
        return 0
    
    # 使用正则表达式提取数字和单位
    match = re.match(r'(\d+(?:\.\d+)?)\s*([亿万]?)', str(value))
    
    if match:
        number, unit = match.groups()
        number = float(number)
        
        if unit == '亿':
            number *= 1e8
        elif unit == '万':
            number *= 1e4
        
        return int(number)
    else:
        return 0
```

```python
from snownlp import SnowNLP
import pandas as pd

# 假设 df_all_weibo_content 是你的数据帧，'微博内容' 是微博内容列
df_all_weibo_content['score'] = df_all_weibo_content['user_self_content'].apply(lambda x: SnowNLP(x).sentiments)
df_all_weibo_content
```

```python
# 定义判断情感标签的函数
def get_sentiment_label(score, threshold_pos=0.6, threshold_neg=0.4):
    if score > threshold_pos:
        return 'positive'
    elif score < threshold_neg:
        return 'negative'
    else:
        return 'neutral'

# # 应用函数生成情感标签列
df_all_weibo_content['label'] = df_all_weibo_content['score'].apply(get_sentiment_label)
df_all_weibo_content.to_csv("df_weibo_post.csv", index=False, encoding='utf-8-sig')
```

```python
import jieba
from collections import Counter
from pyecharts.charts import WordCloud
import pyecharts.options as opts
from pyecharts.globals import SymbolType
```

```python
# 假设 df_all_weibo_content 是你的数据帧，'微博内容' 是微博内容列
text_data = " ".join(df_all_weibo_content['user_self_content'].astype(str))

# 读取中文停用词文件
with open("cn_stopwords.txt", "r", encoding="utf-8") as f:
    stop_words = set(f.read().splitlines())

# 要添加的标点符号、英文字母、特殊符号、表情符等
additional_stop_words = {',', '.', '!', '?', ';', ':', '-', '_', '+', '=', '@', '#', '$', '%', '^', '&', '*', '(', ')', '[', ']', '{', '}', '<', '>', '/', '\\', '|', '`', '~', "'", '"', '😊', '😂', '😁', ...}

# 合并停用词列表
stop_words.update(additional_stop_words)

remove_words = {'a', 'b', 'c','t','b','k','h','|','/','n','→','↑','【','】','~','(',')','!','！','t','s','←','=','#','...','.','·','N','核','污水','日本','核污水','核','日','本','水','水污染','水污染问题','水污染治理','水污染治理问题','水污染治理方案','水污染治理方案问题','水污染治理','核废水'}
stop_words.difference_update(remove_words)

                
# 使用 jieba 进行分词，过滤停用词
words = [word for word in jieba.lcut(text_data) if word not in stop_words]

# 统计词频
word_counts = Counter(words)

# 构建词云图
wordcloud = (
    WordCloud()
    .add("", word_counts.items(), word_size_range=[30, 100], shape="circle")
    .set_global_opts(title_opts=opts.TitleOpts(title="微博核污水内容词云图"))
)

# 生成 HTML 文件（可以使用 render_notebook() 在 Jupyter Notebook 中显示）
wordcloud.render("wordcloud.html")
```
