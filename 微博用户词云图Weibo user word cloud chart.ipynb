```python
import pandas as pd
```

```python
df_weibo_user_info = pd.read_csv("微博用户原始信息-处理.csv")
df_weibo_user_info
```

```python
df_weibo_post = pd.read_csv('核污水微博贴文原始数据.csv')
df_weibo_post
```

```python
df_me = pd.merge(left=df_weibo_post,right=df_weibo_user_info,left_on=['user_name','user_name_href'],right_on=['user_name','user_name_href'])
df_me
```

```python
df_me_gender_content = df_me[['user_name','gender','user_self_content']]
df_me_gender_content
```

```python
df_me_gender_content['user_self_content'] = df_me_gender_content['user_self_content'].str.replace('​','')
df_me_gender_content
```

```python
df_me_gender_content_man = df_me_gender_content[df_me_gender_content['gender'] == '男']
df_me_gender_content_man
```

```python
df_me_gender_content_woman = df_me_gender_content[df_me_gender_content['gender'] == '女']
df_me_gender_content_woman
```

```python
import jieba
from collections import Counter
from pyecharts.charts import WordCloud
import pyecharts.options as opts
```

```python
# 假设 df_all_weibo_content 是你的数据帧，'微博内容' 是微博内容列
text_data = " ".join(df_me_gender_content_man['user_self_content'].astype(str))

# 读取中文停用词文件
with open("cn_stopwords.txt", "r", encoding="utf-8") as f:
    stop_words = set(f.read().splitlines())

# 要添加的标点符号、英文字母、特殊符号、表情符等
additional_stop_words = {',', '.', '!', '?', ';', ':', '-', '_', '+', '=', '@', '#', '$', '%', '^', '&', '*', '(', ')', '[', ']', '{', '}', '<', '>', '/', '\\', '|', '`', '~', "'", '"', '😊', '😂', '😁'}

# 合并停用词列表
stop_words.update(additional_stop_words)

remove_words = {'a', 'b', 'c','t','b','k','h','|','/','n','→','↑','【','】','~','(',')','!','！','t','s','←','=','#','...','.','·','N','核','污水','日本','核污水','核','日','本','水','水污染','水污染问题','水污染治理','水污染治理问题','水污染治理方案','水污染治理方案问题','水污染治理','核废水'}
stop_words.update(remove_words)

                
# 使用 jieba 进行分词，过滤停用词
words = [word for word in jieba.lcut(text_data) if word not in stop_words]

# 统计词频
word_counts = Counter(words)

# 构建词云图
wordcloud = (
    WordCloud()
    .add("", word_counts.items(), word_size_range=[30, 100], shape="circle")
    .set_global_opts(title_opts=opts.TitleOpts(title="微博核污水内容词云图-男性"))
)

# 生成 HTML 文件（可以使用 render_notebook() 在 Jupyter Notebook 中显示）
wordcloud.render("wordcloud_man.html")
```
