```python
df_weibo_post = pd.read_csv('æ ¸æ±¡æ°´å¾®åšè´´æ–‡åŸå§‹æ•°æ®.csv')
df_weibo_post
```

```python
df_me = pd.merge(left=df_weibo_post,right=df_weibo_user_info,left_on=['user_name','user_name_href'],right_on=['user_name','user_name_href'])
df_me
```

```python
df_me_gender_content = df_me[['user_name','gender','user_self_content']]
df_me_gender_content
```

```python
df_me_gender_content['user_self_content'] = df_me_gender_content['user_self_content'].str.replace('â€‹','')
df_me_gender_content
```

```python
df_me_gender_content_man = df_me_gender_content[df_me_gender_content['gender'] == 'ç”·']
df_me_gender_content_man
```

```python
df_me_gender_content_woman = df_me_gender_content[df_me_gender_content['gender'] == 'å¥³']
df_me_gender_content_woman
```

```python
import jieba
from collections import Counter
from pyecharts.charts import WordCloud
import pyecharts.options as opts
```

```python
# å‡è®¾ df_all_weibo_content æ˜¯ä½ çš„æ•°æ®å¸§ï¼Œ'å¾®åšå†…å®¹' æ˜¯å¾®åšå†…å®¹åˆ—
text_data = " ".join(df_me_gender_content_man['user_self_content'].astype(str))

# è¯»å–ä¸­æ–‡åœç”¨è¯æ–‡ä»¶
with open("cn_stopwords.txt", "r", encoding="utf-8") as f:
    stop_words = set(f.read().splitlines())

# è¦æ·»åŠ çš„æ ‡ç‚¹ç¬¦å·ã€è‹±æ–‡å­—æ¯ã€ç‰¹æ®Šç¬¦å·ã€è¡¨æƒ…ç¬¦ç­‰
additional_stop_words = {',', '.', '!', '?', ';', ':', '-', '_', '+', '=', '@', '#', '$', '%', '^', '&', '*', '(', ')', '[', ']', '{', '}', '<', '>', '/', '\\', '|', '`', '~', "'", '"', 'ğŸ˜Š', 'ğŸ˜‚', 'ğŸ˜'}

# åˆå¹¶åœç”¨è¯åˆ—è¡¨
stop_words.update(additional_stop_words)

remove_words = {'a', 'b', 'c','t','b','k','h','|','/','n','â†’','â†‘','ã€','ã€‘','~','(',')','!','ï¼','t','s','â†','=','#','...','.','Â·','N','æ ¸','æ±¡æ°´','æ—¥æœ¬','æ ¸æ±¡æ°´','æ ¸','æ—¥','æœ¬','æ°´','æ°´æ±¡æŸ“','æ°´æ±¡æŸ“é—®é¢˜','æ°´æ±¡æŸ“æ²»ç†','æ°´æ±¡æŸ“æ²»ç†é—®é¢˜','æ°´æ±¡æŸ“æ²»ç†æ–¹æ¡ˆ','æ°´æ±¡æŸ“æ²»ç†æ–¹æ¡ˆé—®é¢˜','æ°´æ±¡æŸ“æ²»ç†','æ ¸åºŸæ°´'}
stop_words.update(remove_words)

                
# ä½¿ç”¨ jieba è¿›è¡Œåˆ†è¯ï¼Œè¿‡æ»¤åœç”¨è¯
words = [word for word in jieba.lcut(text_data) if word not in stop_words]

# ç»Ÿè®¡è¯é¢‘
word_counts = Counter(words)

# æ„å»ºè¯äº‘å›¾
wordcloud = (
    WordCloud()
    .add("", word_counts.items(), word_size_range=[30, 100], shape="circle")
    .set_global_opts(title_opts=opts.TitleOpts(title="å¾®åšæ ¸æ±¡æ°´å†…å®¹è¯äº‘å›¾-ç”·æ€§"))
)

# ç”Ÿæˆ HTML æ–‡ä»¶ï¼ˆå¯ä»¥ä½¿ç”¨ render_notebook() åœ¨ Jupyter Notebook ä¸­æ˜¾ç¤ºï¼‰
wordcloud.render("wordcloud_man.html")
```
